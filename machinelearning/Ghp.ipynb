```python
from sklearn import tree
```

```python
features = [[130], [140], [150], [170]]
```

```python
labels = [0, 0, 1, 1]
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
prediction = clf.predict([[10]])
```

```python
print("Predicción:", prediction)
```
